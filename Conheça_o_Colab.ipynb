### 1. Estruturas de Dados e Classes Base

```
class Produto:
    def __init__(self, id_item, nome, preco):
        self.id_item = id_item
        self.nome = nome
        self.preco = preco

    def __repr__(self):
        return f"[ID: {self.id_item}] {self.nome} - R$ {self.preco:.2f}"

# User Story 1: Tabela Hash para Busca Instantânea
class TabelaHashCardapio:
    def __init__(self, tamanho=100):
        self.tamanho = tamanho
        self.tabela = [[] for _ in range(tamanho)]

    def _hash(self, chave):
        return hash(chave) % self.tamanho

    def inserir(self, produto):
        indice = self._hash(produto.id_item)
        for i, item in enumerate(self.tabela[indice]):
            if item.id_item == produto.id_item:
                self.tabela[indice][i] = produto
                return
        self.tabela[indice].append(produto)

    def buscar(self, id_item):
        indice = self._hash(id_item)
        for item in self.tabela[indice]:
            if item.id_item == id_item:
                return item
        return None

# User Story 4: Árvore Binária de Busca para Relatórios de Preço
class NodoBST:
    def __init__(self, produto):
        self.produto = produto
        self.esquerda = None
        self.direita = None

class BSTPrecos:
    def __init__(self):
        self.raiz = None

    def inserir(self, produto):
        if not self.raiz:
            self.raiz = NodoBST(produto)
        else:
            self._inserir_recursivo(self.raiz, produto)

    def _inserir_recursivo(self, nodo, produto):
        if produto.preco < nodo.produto.preco:
            if nodo.esquerda is None:
                nodo.esquerda = NodoBST(produto)
            else:
                self._inserir_recursivo(nodo.esquerda, produto)
        else:
            if nodo.direita is None:
                nodo.direita = NodoBST(produto)
            else:
                self._inserir_recursivo(nodo.direita, produto)

    def em_ordem(self, nodo, lista):
        if nodo:
            self.em_ordem(nodo.esquerda, lista)
            lista.append(nodo.produto)
            self.em_ordem(nodo.direita, lista)
        return lista
```

### 2. Implementação do Motor Delivery CLI

```
from collections import deque

class MotorDelivery:
    def __init__(self):
        self.cardapio = TabelaHashCardapio()
        self.arvore_precos = BSTPrecos()
        self.carrinho = [] # Pilha para User Story 2
        self.cozinha = deque() # Fila para User Story 3
        self._setup_inicial()

    def _setup_inicial(self):
        itens = [
            Produto("101", "X-Burguer", 25.0),
            Produto("102", "Batata Frita", 15.0),
            Produto("103", "Refrigerante", 8.0),
            Produto("104", "Pizza Individual", 35.0)
        ]
        for p in itens:
            self.cardapio.inserir(p)
            self.arvore_precos.inserir(p)

    def menu_principal(self):
        while True:
            print("\n🛵 MOTOR DELIVERY CLI")
            print("="*40)
            print("1. [Cliente] Ver Cardápio e Buscar Item")
            print("2. [Cliente] Adicionar Item ao Carrinho")
            print("3. [Cliente] Desfazer Último Item do Carrinho")
            print("4. [Cliente] Finalizar Pedido (Enviar para Cozinha)")
            print("5. [Cozinha] Processar/Despachar Próximo Pedido")
            print("6. [Gerente] Ver Relatório de Preços (Árvore)")
            print("0. Sair")

            opcao = input("Escolha uma opção: ")

            if opcao == '1':
                id_busca = input("Digite o ID do item: ")
                item = self.cardapio.buscar(id_busca)
                print(f"\nResultado: {item}" if item else "\nItem não encontrado.")

            elif opcao == '2':
                id_add = input("ID do item para o carrinho: ")
                item = self.cardapio.buscar(id_add)
                if item:
                    self.carrinho.append(item)
                    print(f"{item.nome} adicionado!")
                else: print("ID inválido.")

            elif opcao == '3':
                if self.carrinho:
                    removido = self.carrinho.pop()
                    print(f"Ação desfeita: {removido.nome} removido.")
                else: print("Carrinho vazio.")

            elif opcao == '4':
                if self.carrinho:
                    pedido = list(self.carrinho)
                    self.cozinha.append(pedido)
                    self.carrinho.clear()
                    print("Pedido enviado para a cozinha!")
                else: print("Carrinho vazio.")

            elif opcao == '5':
                if self.cozinha:
                    pedido_pronto = self.cozinha.popleft()
                    print(f"Processando pedido: {pedido_pronto}")
                else: print("Sem pedidos pendentes.")

            elif opcao == '6':
                relatorio = self.arvore_precos.em_ordem(self.arvore_precos.raiz, [])
                print("\nRelatório de Preços (Crescente):")
                for p in relatorio: print(p)

            elif opcao == '0': break
            else: print("Opção inválida.")

# Para iniciar o sistema:
# motor = MotorDelivery()
# motor.menu_principal()
```

```
motor = MotorDelivery()
motor.menu_principal()
```
